```python
explain = {
    "explain":{
        "find":"movies",
        "filter":{"tomatoes.viewer.rating":3.6}
    },
    "verbosity":"executionStats" 
}
new.command(explain)
```

```python
def distilled_explain(explain_output):#limiting explain output and indexing that to obtain perfomance
    return {
        'executionTimeMillis': explain_output['executionStats']['executionTimeMillis'],
        'totalDocsExamined'  : explain_output['executionStats']['totalDocsExamined'],
        'nReturned'          : explain_output['executionStats']['nReturned']
    }
```

```python
filter = new.movies.find({"tomatoes.viewer.rating":3.6}).explain()
```

```python
print(distilled_explain(filter))
```
